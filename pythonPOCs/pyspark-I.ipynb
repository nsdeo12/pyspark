```python
############################################################
########################## Using iloc ######################
############################################################

print("/////////====coulumns======\\\\\\\\\\\\\\\\\n")
print(df)
print("\n/////////=====Take 5 columns=====\\\\\\\\\\\\\\\\\n")
print(df['Name'][0:5])
print("\n/////////=====Take a column=====\\\\\\\\\\\\\\\\\n")
print(df.Name)
print("\n/////////=====Take multiple columns=====\\\\\\\\\\\\\\\\\n")
print(df[['Name','Type 1','HP','Attack']])
print("\n/////////=====integer location=====\\\\\\\\\\\\\\\\\n")
print("=======get the 0th row of the dataframe>>>>>>>>>>>")
print(df.iloc[0])  
print("=======get the rows from 0 to 2>>>>>>>>>>")
print(df.iloc[0:2]) 
print("=======get the rows at 2,1 location>>>>>>>")
print(df.iloc[2,1])
print("========iterate through rows>>>>>>>>")
for index,row in df.iterrows():
    print(index,row[2])

```

```python
############################################################
########################## Sorting    ######################
############################################################
df.sort_values(['Name'],ascending=True)

print("==============sort on two columns>>>>>>>>>>")
df.sort_values(['Name','Attack'],ascending=[1,0])
```

```python
############################################################
########################## Using loc ######################
############################################################
print(df.loc[df["Name"]=="Bulbasaur"])

print("========Find total of different columns>>>>>>>>>")
df['Total']=df['HP']+df['Attack']+df['Defense']+df['Sp. Atk']+df['Sp. Def']+df['Speed']
print(df['Total'])
```
