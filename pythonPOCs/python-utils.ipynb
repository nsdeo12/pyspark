```python
# prime number and composite number
def testPrime(n):
    flag=False
    for i in range(2,n):
        if n%i==0:
            flag=True
            break
    if(flag==True):
        print("Its a composite number")
    else:
        print("Its a prime number")
testPrime(5)                   
```
